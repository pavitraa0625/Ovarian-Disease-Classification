```python
from ultralytics import YOLO

# Dataset directory (ImageFolder format)
DATA_PATH = "/content/drive/MyDrive/MMOTU_splits_full"

# 🔑 Correct checkpoint path (based on your Drive)
LAST_MODEL = "/content/drive/MyDrive/YOLO_MMOTU_RESULTS/yolov8_cls_full/weights/last.pt"

model = YOLO(LAST_MODEL)

print("RESUMING YOLOv8-CLS TRAINING FROM yolov8_cls_full...\n")

model.train(
    data=DATA_PATH,
    epochs=50,              # total epochs
    imgsz=224,
    batch=32,

    lr0=1e-3,
    optimizer="AdamW",
    weight_decay=1e-4,

    project="/content/drive/MyDrive/YOLO_MMOTU_RESULTS",
    name="yolov8_cls_full",  # MUST match folder name

    resume=True,            # 🔑 resume training
    device="cpu",
    workers=0
)
```
