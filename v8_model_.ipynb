```python
from ultralytics import YOLO

DATA_PATH = "/content/drive/MyDrive/MMOTU_splits_full"

model_v8 = YOLO("yolov8n-cls.pt")

model_v8.train(
    data=DATA_PATH,
    epochs=50,
    imgsz=224,
    batch=32,

    project="/content/drive/MyDrive/YOLO_MMOTU_RESULTS",
    name="yolov8_cls_full",   # NEW folder (do not overwrite old)

    # 🔹 FORCE FULL LOGGING
    save=True,
    save_period=1,            # save every epoch
    plots=True,               # VERY IMPORTANT (results.png)
    verbose=True,

    # 🔹 FORCE VALIDATION ARTIFACTS
    val=True,

    # 🔹 Augmentations (same as v11 & v12)
    fliplr=0.5,
    flipud=0.3,
    degrees=10,
    translate=0.1,
    scale=0.5,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4
)
```
