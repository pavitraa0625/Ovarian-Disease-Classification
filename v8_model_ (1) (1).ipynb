```python
from ultralytics import YOLO

DATA_PATH = "/content/drive/MyDrive/MMOTU_splits_full"

model_v8 = YOLO("yolov8n-cls.pt")

model_v8.train(
    data=DATA_PATH,
    epochs=50,
    imgsz=224,
    batch=32,

    project="/content/drive/MyDrive/YOLO_MMOTU_RESULTS",
    name="yolov8_cls_full",   # NEW folder (do not overwrite old)

    # 🔹 FORCE FULL LOGGING
    save=True,
    save_period=1,            # save every epoch
    plots=True,               # VERY IMPORTANT (results.png)
    verbose=True,

    # 🔹 FORCE VALIDATION ARTIFACTS
    val=True,

    # 🔹 Augmentations (same as v11 & v12)
    fliplr=0.5,
    flipud=0.3,
    degrees=10,
    translate=0.1,
    scale=0.5,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4
)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import os

# Construct the path to the results.csv file
results_dir = "/content/drive/MyDrive/YOLO_MMOTU_RESULTS/yolov8_cls_full2" # This should match the output from the training cell
results_path = os.path.join(results_dir, "results.csv")

# Load the results into a DataFrame
df = pd.read_csv(results_path)

# Adjust epoch column if necessary (sometimes it's 0-indexed)
df["epoch"] = df["epoch"] + 1

plt.figure(figsize=(10, 6))
plt.plot(df["epoch"], df["metrics/accuracy_top1"], label="Validation Top1 Accuracy")
plt.plot(df["epoch"], df["train/loss"], label="Train Loss")
plt.xlabel("Epochs")
plt.ylabel("Value")
plt.legend()
plt.title("YOLOv8 Training Loss vs Validation Accuracy")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df["epoch"], df["train/loss"], label="Train Loss")
plt.plot(df["epoch"], df["val/loss"], label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("YOLOv8 Training and Validation Loss over Epochs")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df["epoch"], df["metrics/accuracy_top1"], label="Validation Accuracy (Top-1)")
plt.plot(df["epoch"], df["metrics/accuracy_top5"], label="Validation Accuracy (Top-5)")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("YOLOv8 Validation Accuracy (Top-1 and Top-5) over Epochs")
plt.legend()
plt.grid(True)
plt.show()
```

```python
final_acc = df["metrics/accuracy_top1"].iloc[-1]
print(f"Final Validation Accuracy (YOLOv8): {final_acc:.4f}")
```
